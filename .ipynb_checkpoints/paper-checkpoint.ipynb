# Get new pages and editors

```python
def GetNewPandE(df):
    dict_new_editors ={}
    dict_new_pages = {}
    for year in np.arange(2000,2020):

        n_new_editors = df[(~df['user'].isin(df[df['timestamp'].dt.year < year]['user'].unique())) &(df['timestamp'].dt.year >= year)&
             (df['timestamp'].dt.year < year+1)]['user'].nunique()
        new_pages = list(df_edits[(~df['name'].isin(df[df['timestamp'].dt.year < year]['name'].unique())) &(df['timestamp'].dt.year >= year)&
             (df['timestamp'].dt.year < year+1)]['name'].unique())
        dict_new_editors[year] = {'new_editors':n_new_editors}
        dict_new_pages[year] = new_pages
    df_new_editors = pd.DataFrame.from_dict(dict_new_editors,orient='index')
    return df_new_editors,dict_new_pages
```

```python
class AnyObjectHandler(HandlerBase):
    def create_artists(self, legend, orig_handle,
                       x0, y0, width, height, fontsize, trans):
        l1 = plt.Line2D([x0,y0+width], [0.9*height,0.9*height],
                           marker='o',markerfacecolor=orig_handle[1],color=orig_handle[0])
        l2 = plt.Line2D([x0,y0+width], [0.1*height,0.1*height],linestyle=orig_handle[2], color=orig_handle[0])
        return [l1, l2]
```

```python
def GETP(r,n):
    t = (r*sqrt(n-2))/(sqrt(1-(r**2)))
    
    pval = sp.t.sf(np.abs(t), n-2)*2
    return pval
```

```python
def StyleMyfig(a,x_label,y_label,stylefonts):
    a.set_xlabel(x_label,fontsize=stylefonts['labels_font'],labelpad=8)
    a.set_ylabel(y_label,fontsize=stylefonts['labels_font'],labelpad=8)
    a.tick_params(axis='both', which='major', labelsize=stylefonts['ticks_font'],direction='in',pad=7,width=1.5,length=5)
    plt.xticks(rotation=0)

    #a.tick_params(axis='both', which='minor', labelsize=0,direction='in',pad=7,width=0)
    

    for tick in a.xaxis.get_major_ticks():
        tick.label1.set_horizontalalignment('center')
        tick.label1.set_rotation(0)
    [j.set_linewidth(1.5) for i,j in a.spines.items()]
```

## Figure 3: Correlation between attention on Wikipedia and market performance

```python
f3_btc_part = df[(df['name']=='Bitcoin')&(df['time'] >= '2015-07-01')][['views_page','price','time','market_share']].resample('W',on='time').agg(
    {'views_page':'sum', 'price':'last'}).dropna()
```

```python
f3_btc_part['views_page'] = f3_btc_part['views_page']/1000
```

```python
f3_btc_part.rename(columns={'views_page':'wikipedia views (in thousands)'},inplace=True)
```

```python
views_means = df.dropna(subset=['views_share','volume_share']).groupby('name').mean()['views_share']
price_means = df.dropna(subset=['price','views_share']).groupby('name').mean()['price']
edits_means = df.dropna(subset=['edits_share','volume_share']).groupby('name').mean()['edits_share']
volume_means = df.dropna(subset=['volume_share','volume_share']).groupby('name').mean()['volume_share']
market_shares_means = df.dropna(subset=['market_share','volume_share']).groupby('name').mean()['market_share']
```

```python
market_shares_means =  market_shares_means[market_shares_means.index!='Filecoin']  
# Filecoin is a future so no market share listed in coinmarket cap
```

```python
slope_m_v,intercept_m_v,rvalue_m_v,pvalue_m_v ,stderr_m_v = sp.linregress(np.log10(views_means[views_means.index!='Filecoin']),np.log10(market_shares_means))
slope_m_e,intercept_m_e,rvalue_m_e,pvalue_m_e ,stderr_m_e = sp.linregress(np.log10(edits_means[edits_means.index!='Filecoin']),np.log10(market_shares_means))
```

# Evolution of cryptocurrency pages

## Figure 4: Example of edit histories

```python
diff_bet_rev = df_edits.groupby('name').apply(lambda a: a.sort_values('timestamp')['timestamp'].diff().dt.days).dropna()
```

```python
diff_bet_rev_btc =  diff_bet_rev.loc['Bitcoin']
diff_bet_rev_dash =  diff_bet_rev.loc['Dash']
```

```python
hist_diff,bin_diff = np.histogram(diff_bet_rev.values+1,density=True,bins=np.logspace(np.log10(1),np.log10(max(diff_bet_rev.values+1)),12))
hist_diff_btc,bin_diff_btc = np.histogram(diff_bet_rev_btc.values+1,density=True,bins=np.logspace(np.log10(1),np.log10(max(diff_bet_rev_btc.values+1)),12))
hist_diff_dash,bin_diff_dash = np.histogram(diff_bet_rev_dash.values+1,density=True,bins=np.logspace(np.log10(1),np.log10(max(diff_bet_rev_dash.values+1)),12))
```

```python
bin_diff_center = (bin_diff[:-1] + bin_diff[1:]) / 2
bin_diff_center_btc = (bin_diff_btc[:-1] + bin_diff_btc[1:]) / 2
bin_diff_center_dash = (bin_diff_dash[:-1] + bin_diff_dash[1:]) / 2
```

```python
x_diff,y_diff = zip(*[(i,j) for (i,j) in zip(bin_diff_center,hist_diff) if j != 0])  
x_diff_btc,y_diff_btc = zip(*[(i,j) for (i,j) in zip(bin_diff_center_btc,hist_diff_btc) if j != 0])  
x_diff_dash,y_diff_dash = zip(*[(i,j) for (i,j) in zip(bin_diff_center_dash,hist_diff_dash) if j != 0])  
```

```python
fit_dist_sec_bet_rev = sp.linregress(np.log10(x_diff[1:]),np.log10(y_diff[1:]))
fit_dist_sec_bet_rev_btc = sp.linregress(np.log10(x_diff_btc[1:]),np.log10(y_diff_btc[1:]))
fit_dist_sec_bet_rev_dash = sp.linregress(np.log10(x_diff_dash[1:]),np.log10(y_diff_dash[1:]))
```

```python
fig,axs = plt.subplots(2,2,figsize=(7.1,3.7))
gs = gridspec.GridSpec(2, 2)
ax1 = plt.subplot(gs[:, 0])
ax2 = plt.subplot(gs[0, -1])
ax3 = plt.subplot(gs[1, -1])
axs = [ax1,ax2,ax3]
########################################################Figure A##########################################################################

ax1.loglog(x_diff_btc,y_diff_btc,color=colors[2],lw=3,marker='o')

label_btc = 'Bitcoin \n $x^{{%0.02f}\pm{%0.02f}}$'%(fit_dist_sec_bet_rev_btc[0],fit_dist_sec_bet_rev_btc[4])
ax1.loglog(x_diff_dash,y_diff_dash,color=colors[2],lw=3,marker='o',markerfacecolor='white')

label_dash = label='Dash \n $x^{{%0.02f}\pm{%0.02f}}$'%(fit_dist_sec_bet_rev_dash[0],fit_dist_sec_bet_rev_dash[4])

ax1.loglog(x_diff_btc,[10**fit_dist_sec_bet_rev_btc[1]*i**fit_dist_sec_bet_rev_btc[0] for i in x_diff_btc],color=colors[2],ls='--',
                                 label='$x^{{%0.02f}\pm{%0.02f}}$'%(fit_dist_sec_bet_rev_btc[0],fit_dist_sec_bet_rev_btc[4]))

ax1.loglog(x_diff_dash,[10**fit_dist_sec_bet_rev_dash[1]*i**fit_dist_sec_bet_rev_dash[0] for i in x_diff_dash],color=colors[2],ls='-.',
                                 label='$x^{{%0.02f}\pm{%0.02f}}$'%(fit_dist_sec_bet_rev_dash[0],fit_dist_sec_bet_rev_dash[4]))
plot_ls =['--','-.']
markers = [colors[2],'white']
sty = [(colors[2],markers[i],plot_ls[i]) for i in np.arange(2)]
legend1 = ax1.legend(sty,[label_btc,label_dash],loc=1,fontsize=stylefonts['legend_font'],
                     handlelength=2,handletextpad=1,labelspacing=0.4,borderpad=0.4,
           handler_map={tuple: AnyObjectHandler()})
ax1.add_artist(legend1)
StyleMyfig(ax1,'time between edits (in days)','pdf',stylefonts)

############################################################################################################################################
###################################################Figure B################################################################################
i =1     
letters=['A','B','C']
for page_name, df_now in df_edits[df_edits['name'].isin(['Bitcoin','Dash'])].groupby('name'):
    a = axs[i]
    dates = pd.date_range(df_now.resample('D',on='timestamp').count().index.min(),
    df_now.resample('D',on='timestamp').count().index.max(),freq='D')
    edits_happ = df_now.resample('D',on='timestamp').count()
    edits_happ = edits_happ[edits_happ['name'] > 0]
    ind_edit = [i for i,x in enumerate(dates.isin(edits_happ.index)) if x == True]        
    a.vlines(x =ind_edit[1:],ymin=0,ymax=1,lw=0.2,alpha=0.2)
        
    a.set_yticks([])
    a.annotate(letters[i],  xy=(1.02, 0.5), xycoords='axes fraction',weight='bold',fontsize=18)
    i += 1
    
    StyleMyfig(a,'days since creation','%s edits'%page_name,stylefonts)


ax1.annotate('A',  xy=(-0.08, 1), xycoords='axes fraction',weight='bold',fontsize=20)

plt.tight_layout()
fig.savefig('fig_4.eps',bbox_inches='tight')
```

## Figure 6: Uneven distribution of contributions of Wikipedia editors

```python
mini_df_edits = df_edits[['user','timestamp']]
```

```python
share_in_edits = ((mini_df_edits.groupby('user').count().sort_values('timestamp',ascending=False)['timestamp']/mini_df_edits.count()['timestamp'])*100).to_frame('share in edits').set_index(np.arange(
                                        1,mini_df_edits['user'].nunique()+1))
```

```python
def GetFitPoint(share_edits):
    bests = [np.inf,0]

    for i in np.arange(share_edits.count()/2,share_edits.count(),dtype=int):
        slope_share_e,intercept_share_e,rvalue_share_e,pvalue_share_r,stderr_share_e = sp.linregress(np.log10(share_edits.index[:i]),
                                                                np.log10(share_edits['share in edits'].iloc[:i]))
        if stderr_share_e < bests[0]:
            bests[0] = stderr_share_e
            bests[1] = i
    return bests
```

```python
bests = GetFitPoint(share_in_edits)
```

### Users share of edits in different years

```python
per_year_cont = df_edits[df_edits['timestamp'].dt.year >= 2012].resample('Y',on='timestamp').apply(lambda a: 
                                        (a.groupby('user').count()['name']/a.count()['name']).sort_values(ascending=False)).reset_index()
```

```python
time_diff_bet_revisions = df_edits.groupby('user').apply(lambda a: a[['name','timestamp']].groupby('name').diff()['timestamp'].dt.seconds)
```

```python
compresed_df_edits = df_edits[df_edits.index.isin(
    time_diff_bet_revisions[(time_diff_bet_revisions.isna())|(time_diff_bet_revisions > 3600 )].index.get_level_values(1))]
```

```python
compressed_share = (compresed_df_edits.groupby('user').count()['name']/compresed_df_edits.count()['name']).sort_values(ascending=False)
```

```python
fig,axs = plt.subplots(1,2,figsize = (10,4))
lines=[]
for grp_name, grp in per_year_cont.groupby('timestamp'):
    x = np.arange(1,(grp.count()['name'])+1)
    slope_y,intercept_y,rvalue_y,pvalue_y,stderr_y = sp.linregress(np.log10(x),np.log10(grp['name']))
    line = axs[0].loglog(x,grp['name'],color=colors[3],lw=3,label = '%s \n $r^{%0.2f \pm %0.2f}$'%(grp_name.year,slope_y,stderr_y))
    lines.append(line[0])
first_legend = axs[0].legend(handles=lines[:4], loc=1,ncol=2,fontsize=10)

axs[0].add_artist(first_legend)
axs[0].legend(handles=lines[4:], loc=3,ncol=2,fontsize=11)

StyleMyfig(axs[0],'rank','share in edits',stylefonts)
axs[0].annotate('A',  xy=(-0.04, 1.05), xycoords='axes fraction',weight='bold',fontsize=20)
axs[1].annotate('B',  xy=(-0.04, 1.05), xycoords='axes fraction',weight='bold',fontsize=20)
######################################################################################################################################
x_c = np.arange(1,compressed_share.count()+1)
slope_c,intercept_c,rvalue_c,pvalue_c,stderr_c = sp.linregress(np.log10(x_c),np.log10(compressed_share.values))

axs[1].loglog(x_c,compressed_share.values,color=colors[3],lw=3,label='%s \n $r^{%0.2f \pm %0.2f}$'%('one hour',slope_c,stderr_c))

StyleMyfig(axs[0],'rank','share in edits',stylefonts)

axs[1].legend(fontsize=stylefonts['legend_font'])
#############################################################################################

plt.tight_layout()
#fig.savefig('../../wikipedia_cryptos/paper_figures/fig_AP1_share_fit_time.pdf',bbox_inches='tight')
```
